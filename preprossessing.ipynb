```python
import pandas as pd
```

```python
twitter_dataset=pd.read_csv("twcs.csv")
```

```python
twitter_dataset
```

```python
#Cleaning
import re
```

```python
# Clean the 'text' column : removing ponctuation
twitter_dataset["text_cleaned"] = twitter_dataset["text"].apply(lambda x: re.sub(r'[^\w\s]', '', str(x)))
# Display the modified DataFrame
twitter_dataset
```

```python
#Normalize the text_cleaned column:

twitter_dataset["text_cleaned"] = twitter_dataset["text_cleaned"].str.lower()
twitter_dataset
```

```python
#Selecting important columns:
selected_cols =['tweet_id','text','text_cleaned']
twitter_dataset_cleaned = twitter_dataset.loc[:, selected_cols]
twitter_dataset_cleaned
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
#Tokinazation : spliting the string into a list of words
from nltk.tokenize import word_tokenize
from nltk import sent_tokenize

twitter_dataset_cleaned['text_cleaned'] = twitter_dataset_cleaned['text_cleaned'].apply(lambda x: word_tokenize(str(x)))
twitter_dataset_cleaned
```

```python
#Removing stop_words : Removing stop words involves eliminating words that do not convey meaningful content.
from nltk.corpus import stopwords

nltk.download('stopwords')

# Get the English stop words

stop_words = stopwords.words('english')


twitter_dataset_cleaned['text_cleaned'] = twitter_dataset_cleaned['text_cleaned'].apply(lambda sentence_tokens: [word for word in sentence_tokens if word not in stop_words ])

twitter_dataset_cleaned
```

```python
#To remove the rare and frequent words we need to count first the frequency of each word:

from collections import Counter

# Count word frequencies
word_freq = Counter(word for words_list in twitter_dataset_cleaned['text_cleaned'] for word in words_list)
```

```python
total_documents = len(twitter_dataset_cleaned)
extremely_frequent_words = [word for word, freq in word_freq.items() if freq > 0.9 * total_documents]
very_rare_words = [word for word, freq in word_freq.items() if freq < 0.01 * total_documents]
```

```python
def remove_freq_rare_words(words_list, freq_words, rare_words):
    return [word for word in words_list if word not in freq_words and word not in rare_words]

twitter_dataset_cleaned['text_cleaned'] = twitter_dataset_cleaned['text_cleaned'].apply(
    lambda x: remove_freq_rare_words(x, extremely_frequent_words, very_rare_words)
)
twitter_dataset_cleaned
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()


twitter_dataset_cleaned['text_cleaned'] = twitter_dataset_cleaned['text_cleaned'].apply(
    lambda x: [lemmatizer.lemmatize(word) for word in x]
)
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()


twitter_dataset_cleaned['text_cleaned'] = twitter_dataset_cleaned['text_cleaned'].apply(
    lambda x: [stemmer.stem(word) for word in x]
)
```
