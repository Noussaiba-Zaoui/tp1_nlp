```python
model = gensim.models.Word2Vec(
    window=10,#m = 10
    min_count=2,#at least two words in the sentence to do training
    workers=8,#number of threads in my cpu
)

    
```

```python
#building the vocabulary :
#It analyzes the given input data to identify unique words and assign them unique integer indices.
#a progress update will be printed after processing every 1000 texts.
model.build_vocab(processed_text, progress_per=1000)
```

```python
model.train(processed_text, total_examples=model.corpus_count, epochs=model.epochs)
```
